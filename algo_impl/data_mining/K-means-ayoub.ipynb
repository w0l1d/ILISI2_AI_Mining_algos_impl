```python
class K_Means:

    def __init__(self, k=2,centroidV=[], tolerance = 0.001, max_iter = 500):
        self.k = k
        self.centroid = {}
        self.max_iterations = max_iter
        self.tolerance = tolerance
        if len(centroidV) == 0:
            self.centroid = {}
        else:
            self.centroid = {}
        #self.centroid = centroid

    def euclidean_distance(self, point1, point2):
        #return math.sqrt((point1[0]-point2[0])**2 + (point1[1]-point2[1])**2 + (point1[2]-point2[2])**2)   #sqrt((x1-x2)^2 + (y1-y2)^2)
        return np.linalg.norm(point1-point2, axis=0)

    def predict(self,data):
        distances = [np.linalg.norm(data-self.centroids[centroid]) for centroid in self.centroids]
        classification = distances.index(min(distances))
        return classification
    def indexOfElementInTupleNestedList( self,list, element):
        result =np.where(np.all(list == element, axis = 0))
        print(result)
        if len(result[0])>0:
            return int(result[0][0])
        return -1
    def find_element_index(self,nested_list, target):
        for i, element in enumerate(nested_list):
            if isinstance(element, list):
                index = self.find_element_index(element, target)
                if index is not None:
                    return [i] + index
            elif element == target:
                return [i]
        return -1
    def find_array_index_in_dict(self,my_dict, my_array):
        for key, value in my_dict.items():
            if (value == my_array).any():
                return list(my_dict.keys()).index(key)
        return -1

    def fit(self, data):


        if len(self.centroid) == 0:
            self.centroids = {}
            for i in range(self.k):
                self.centroids[i] = data[i]
        else:
            self.centroids = {}
            for i in range(self.k):
                self.centroids[i] = self.centroid[i]

        self.classes=data[:self.k]
        clonedCentroids=self.centroids.copy()
        print("On prend "+ str(self.k) +" objets au hasard :" +  str(list(self.classes)))
        print("\n")
        clonedCentroids= dict(zip(self.centroids, map(lambda x: " C{%s} = %s, M%s = %s" % (str(x+1) , str(self.centroids[x]),str(x+1) , str(self.centroids[x])), self.centroids)))
        print(clonedCentroids)
        self.classes= {}
        for j in range(self.k):
            self.classes[j] = []


        for point in data:
            distances = []
            for index in self.centroids:
                distances.append(self.euclidean_distance(point,self.centroids[index]))
            cluster_index = distances.index(min(distances))
            self.classes[cluster_index].append(point)


        for i in range(self.max_iterations):
            previousClasses= self.classes.copy()
            self.classes = {}
            for j in range(self.k):
                self.classes[j] = []

            for point in data:
                distances = []
                for index in self.centroids:
                    distances.append(self.euclidean_distance(point,self.centroids[index]))
                cluster_index = distances.index(min(distances))
                if( len(previousClasses) >0):
                    prevIndex=self.find_array_index_in_dict(previousClasses, point)
                else:
                    prevIndex=-1
                if len(previousClasses)>0 and prevIndex != -1:
                    if prevIndex !=  cluster_index:
                        previous_index=prevIndex
                        print("\n")
                        print("  dist(%s, M%s) < dist(%s, M%s) ⇒ %s passe dans C%s Tous les autres objets ne bougent pas " % (str(point), str(cluster_index+1) ,str(point),  str(previous_index+1),str(point), str(cluster_index+1)))
                    self.classes[cluster_index].append(point)

            previous = dict(self.centroids)
            for cluster_index in self.classes:
                self.centroids[cluster_index] = np.average(self.classes[cluster_index], axis = 0)
            print("\n")
            clonedCentroids= dict(zip(self.centroids, map(lambda x: " C{%s} = %s, M%s = %s" % (str(x+1) , str(list(self.classes[x])),str(x+1) , str(self.centroids[x])), self.centroids)))
            print(list(clonedCentroids.values()))



            isOptimal = True

            for centroid in self.centroids:
                original_centroid = previous[centroid]
                curr = self.centroids[centroid]
                if np.sum((curr - original_centroid)/original_centroid * 100.0) > self.tolerance:
                    isOptimal = False
            if isOptimal:
                break



def main():
    K=3


    #A=np.array([ [15,16],[12,13],[16,20],[10,11],[11,12.5]])
    #A=np.array([ [1],[2],[3],[6],[7],[8],[9],[13],[15],[17]])
    A=np.array([ [1,1],[2,2],[3,3],[6,6],[7,7],[8,8],[9,9],[13,13],[15,15],[17,17]])
    centroides= np.array([ [11,13],[14,18]])
    centroides= np.array([])
    #k_means = K_Means(K,centroides)
    k_means = K_Means(K,centroides)
    k_means.fit(A)
    centroids = k_means.centroids
    labels = k_means.classes
    print("\n")
    print("Centroides "+ str(centroids) )
    print("\n")
    print("Clusters "+ str(labels) )
    value_to_predict = np.array([12,15])
    print("\n")
    print(" %s sera dans cluster C{%s}"% ( str(value_to_predict) ,str(k_means.predict(value_to_predict)+1) ) )



    # Plotting starts here
#     colors = 10*["r", "g", "c", "b", "k"]
#
#     for centroid in k_means.centroids:
#         plt.scatter(k_means.centroids[centroid][0], k_means.centroids[centroid][1], s = 130, marker = "x")
#
#     for cluster_index in k_means.classes:
#         color = colors[cluster_index]
#         for features in k_means.classes[cluster_index]:
#             plt.scatter(features[0], features[1], color = color,s = 30)
#
if __name__ == "__main__":
    main()
```
